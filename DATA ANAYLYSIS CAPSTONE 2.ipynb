```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Preprocessing:
‚óè Correct the data in the "height" column by replacing it with random
numbers between 150 and 180. Ensure data consistency and integrity
before proceeding with analysis. (1 mark)

```python
data = pd.read_csv("ABC Company.xlsx - Sheet1.csv")
data.head(10)
```

```python
data["Height"] = np.random.randint(150,180 , size = len(data))
data.head(10)
```

```python
### NOW SAVE THE FILE INTO A NEW CSV 
data.to_csv("cleaned_data.csv" , index= False)
```

```python
cleaned = pd.read_csv("cleaned_data.csv")
cleaned.head(20)
```

Determine the distribution of employees across each team and calculate
the percentage split relative to the total number of employees. (2 marks)

```python

t_count = cleaned['Team'].value_counts()
print(t_count)
```

```python
## percentage can be taken by a / total *100

t_count_percentage = (t_count / len(cleaned) ) * 100
print(t_count_percentage.round(2))
```

Segregate employees based on their positions within the company

```python
cleaned.head(3)
```

```python
grouped_position = cleaned.groupby('Position')

for position, group in grouped_position:
    print(f"\nPOSITION: {position}")
    # Print only the important columns
    print(group[['Name', 'Team']].to_string(index=False))
```

Identify the predominant age group among employees. (2 marks)

```python
cleaned.head(10)

age_count = cleaned['Age'].value_counts()
print(age_count)
print()
predominant_group = age_count.idxmax()
print(f'The dominand age group is {predominant_group}')
print()
group_age = age_count.max()
print(f'The num of people in the age group is {group_age}')
print()
print(f'THE MOST DOMINANT AGE GROUP IN THE FILE IS {predominant_group} and the count of the people in this age group  is {group_age}')


```

Discover which team and position have the highest salary expenditure. (2
marks)

```python
cleaned.head(10)
```

```python
most_paid_team_list= cleaned.groupby("Team")["Salary"].sum().sort_values(ascending=False)
print("THE TEAM EXPENDITURE LIST OF TOP 5 IS  : ")
print(most_paid_team_list.head())
print("----------------------------------------------------------")
print()
most_paid_team = most_paid_team_list.idxmax()
print(f'THE MOST PAID TEAM ON THE LIST IS  : {most_paid_team}')
print("----------------------------------------------------------")
print()
most_paid_position_list =  cleaned.groupby('Position')['Salary'].sum().sort_values(ascending=False)
print("THE TOP 5 PAID POSITION ON THE LIST  :")
print(most_paid_position_list.head())
print("----------------------------------------------------------")
print()
most_paid_position = most_paid_position_list.idxmax()
print(f'THE MOST PAID POSITION ON THE LIST IS :  {most_paid_position}')
print()
```

Investigate if there's any correlation between age and salary, and
represent it visually. (2 marks)

```python
corr = cleaned['Age'].corr(cleaned['Salary'])
print(f'THE CORELATION BTW THE AGE AND SALARY IS {corr:.2f}')
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(cleaned['Age'], cleaned['Salary'], edgecolors='w', linewidth=10)
plt.title('Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid(True)
plt.show()
```
